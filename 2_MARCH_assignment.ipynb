ANS = Matplotlib is a comprehensive data visualization library in Python widely used for creating static, interactive, and publication-quality plots. It provides a convenient way to generate various types of plots and charts, allowing users to visualize data in a clear and meaningful manner.

The Pyplot module within Matplotlib provides a MATLAB-like interface to create plots easily. Some commonly used plots that can be generated using Pyplot include:

Line Plot: 

Shows data points connected by straight line segments, suitable for displaying trends over time or continuous data.


Scatter Plot:

Displays individual data points as markers, useful for visualizing relationships between two variables.

Histogram:

Represents the distribution of a continuous variable by dividing data into bins and displaying the frequency of observations in each bin.

Bar Chart:

Uses bars of varying heights to represent categorical data, comparing quantities across different categories.


Pie Chart: Illustrates the proportion or percentage of each category in a dataset, represented as slices of a pie.

ANS = A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional graph. Each dot represents the value of two different variables, one plotted along the x-axis and the other along the y-axis.

The primary purpose of a scatter plot is to show the relationship or correlation between these two variables. It's particularly useful for identifying patterns, trends, or the absence thereof, in datasets.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.title("scatter plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.scatter(x,y,c='g')
plt.show()
```

ANS = The subplot() function in Matplotlib is used to create a grid of plots within a single figure. It allows you to divide the figure into rows and columns and specify the location of a subplot based on these divisions. This function is particularly helpful when you want to display multiple plots or charts simultaneously.

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) #for line one 
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) # for line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5]) # for line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5]) # for line 4
y4 = np.array([200, 350, 250, 550, 450, 150])


plt.figure(figsize=(10, 8))  # Defining figure size

# Subplot 1
plt.subplot(2,2,1)  # 2 rows, 2 columns, subplot 1
plt.plot(x1, y1)
plt.title('Plot 1')


plt.subplot(2, 2, 2)  # 2 rows, 2 columns, subplot 2
plt.plot(x2, y2)
plt.title('Plot 2')

# Subplot 3
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, subplot 3
plt.plot(x3, y3)
plt.title('Plot 3')

# Subplot 4
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, subplot 4
plt.plot(x4, y4)
plt.title('Plot 4')

plt.tight_layout()  # Adjust layout to prevent overlapping


plt.show()
```

ANS = 
A bar plot, also known as a bar chart, is a graphical representation of categorical data using rectangular bars. Each bar represents the value of a particular category, and the length or height of the bar corresponds to the value it represents.

```python
company= np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.xlabel("company")
plt.ylabel("profit")
plt.bar(company,profit,color='#4C225D')
plt.show()
```

```python
# HORIZONTAL bar plot
company= np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.xlabel("company")
plt.ylabel("profit")
plt.barh(company,profit,color='#4C225D')
plt.show()
```

ANS = A box plot, also known as a box-and-whisker plot, is a graphical representation that displays the distribution of a dataset based on its five-number summary: minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum.

Key components of a box plot:

Median (Q2): 
    
   Represents the middle value of the dataset. It is the line inside the box.

Interquartile Range (IQR): 
    
The range between the first quartile (Q1) and the third quartile (Q3), forming the box.

Whiskers:

Lines extending from the box to the minimum and maximum values within a defined range (typically 1.5 times the IQR).

Outliers: Individual data points lying beyond the whiskers, considered as exceptional or unusual values.


Uses of a Box Plot:
    
Visualizing Data Distribution: Box plots provide a clear visual summary of the distribution of numerical data, indicating the spread, central tendency, and skewness.

Identifying Outliers:
    
They help in identifying potential outliers in the dataset, assisting in anomaly detection or understanding the variability in the data.


Comparing Distributions: 

Useful for comparing distributions of multiple datasets or groups, allowing for easy comparison of central tendencies and spreads.


Assessing Symmetry and Skewness: 
    
Box plots visually depict the symmetry or skewness of the data based on the box's position and size

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1, box2]
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of box1 and box2')
plt.grid(True)
plt.show()
```
